```python
#Account Information
account_holder="Dhanya Krishnan"
account_number="234567785"
balance=1000.00
currency="Rs"
```

```python
def deposit(amount):
 global balance
 if amount<=0:
  print("Deposit amount must be positive");
  return balance
 balance+=amount
 print("Deposit Successfull")
 return balance
```

```python
def withdrawal(amount):
 global balance
 if amount <=0:
  print(f"Withdrawal amount not be less than zero")
  return
 elif(amount>balance):
  print(f"Insufficient funds available:{currency}{balance}")
  return
 else:
   balance-=amount
   print("Withdraw successfully")
```

```python
def check_balance():
  print(f"Account Holder:{account_holder}")
  print(f"Account Number:{account_number}")
  print(f"Current Balance:{currency}{balance}")
```
